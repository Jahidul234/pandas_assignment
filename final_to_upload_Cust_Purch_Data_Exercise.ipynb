**10. What is the max, min and average spending?**

```python
max_spending = ecommerce['price(CAD)'].max()
min_spending = ecommerce['price(CAD)'].min()
avg_spending = ecommerce['price(CAD)'].mean()
print(f"Max Spending: {max_spending}")
print(f"Min Spending: {min_spending}")
print(f"Average Spending: {avg_spending}")
```
